### · 자료형 & 기초연산

```python
# 자료형
def convType(x):
    return type(x).__name__

exType = [10, 2, 3.3, 18.5, '안녕', '하세요', True, False];

for cType in exType :
    print(f"{cType} ({convType(cType)})")
```

```python
def dType(x):
    return type(x).__name__

age = 19;
ages = [10, 20, 30, 18.5, '안녕'];

# 제한연령 체크
for check in ages :
    if not isinstance(check, int) : 
        print(f"Type MissMatch -> {check}({dType(check)})")
        continue
    print(
        f"제한연령인 {age}보다 "
        f"{abs(age - check)} 만큼 "
        f"{'작습니다' if check < age else '큽니다'}"
    )
```

### · 인덱싱 & 슬라이싱

```python
# 단순
text = "123456789";
print(f"idx(2) : {text[2]}")
print(f"1step : {text[::2]}")
print(f"idx(2~4) : {text[2:5]}")

# 실사용
def repNum(phone):
    return phone.replace("-", "")

register = ["010-1234-5678", "02-234-5678", "010-9876-5432"]
db = {"01056781234": "헬로키티", "01012345678": "시나모롤", "022345678": "산리오코리아"}
for phone in register :
    key = repNum(phone)

    if key in db:
        name = db[key]
        print(f"{name}({phone[-4:]})님 가입 이력이 있습니다.")
```
